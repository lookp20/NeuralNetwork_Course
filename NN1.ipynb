```python
import numpy as np
```

```python
data = np.loadtxt("iris.csv", delimiter=",")
```

```python
X = data[:, :4]
T = data[:, -1]
```

```python
import matplotlib.pyplot as plt 
```

```python
X_transpose = X.T
```

```python
plt.scatter(X_transpose[0,:], X_transpose[1,:], c = T, cmap = plt.cm.prism)
```

```python
def neuron(X_transpose):
    W = [-0.3, 1]
    treshold = 1.5
    summation = np.dot(W, X_transpose[:2])
    res = summation > treshold
    return res
```

```python
import nnwplot
```

```python
nnwplot.plotTwoFeatures(X_transpose[:2],T,neuron)
```
